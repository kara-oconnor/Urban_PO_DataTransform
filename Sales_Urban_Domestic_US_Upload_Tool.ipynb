```python
# BDay is business day, not birthday...
from pandas.tseries.offsets import BDay

today = datetime.datetime.today()

# QL/ Ship Cancel = UO Cancel(aka date due in blankdf) - BDay(5) 
blankdf['customer_cancel_date'] = pd.to_datetime(blankdf['customer_cancel_date'])
blankdf['cancel_date'] = blankdf['customer_cancel_date'] - BDay(5)



# Subtracts Cencel date by days depending on if weekend or Saturday or Sunday 
blankdf['weekday'] = blankdf['customer_cancel_date'].dt.dayofweek
blankdf.loc[blankdf['weekday'] <= 4, "date_due"] = (blankdf['customer_cancel_date'] - datetime.timedelta(14)) # if weekday
blankdf.loc[blankdf['weekday'] == 5, "date_due"] = (blankdf['customer_cancel_date'] - datetime.timedelta(15)) # if Sat
blankdf.loc[blankdf['weekday'] == 6, "date_due"] = (blankdf['customer_cancel_date'] - datetime.timedelta(16)) # if Sun


# This allows it to revert back to string(uneditable by numbers)

# Would also need to do this for test/change name if we do it in new column 
blankdf['date_due'] = blankdf['date_due'].dt.strftime("%m/%d/%y") 
blankdf['customer_cancel_date'] = blankdf['customer_cancel_date'].dt.strftime("%m/%d/%Y") 
blankdf['cancel_date'] = blankdf['cancel_date'].dt.strftime("%m/%d/%Y")
del blankdf['weekday']
```
